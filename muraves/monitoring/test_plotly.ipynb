```python
import plotly.graph_objects as go

def plot_3d_transparent_planes():
    fig = go.Figure()

    # Primi tre piani (Station 1, Station 2, Station 3)
    planes = [
        {"name": "S0", "x": -0.26, "y": 0, "z": 0, "size": 1, "height": 1, "color": "rgba(180, 180, 220, 0.5)"},  # Grigio trasparente
        {"name": "S1", "x": 0, "y": 0, "z": 0, "size": 1, "height": 1, "color": "rgba(180, 180, 220, 0.5)"},    # Blu chiaro trasparente
        {"name": "S2", "x": 0.25, "y": 0, "z": 0, "size": 1, "height": 1, "color": "rgba(180, 180, 220, 0.5)"}, # Rosa chiaro trasparente
    ]

    # Aggiungi i piani come superfici 2D sul piano yz (senza spessore in x)
    for plane in planes:
        half_size = plane["size"] / 2
        fig.add_trace(go.Surface(
            x=[[plane["x"], plane["x"]], [plane["x"], plane["x"]]],  # Stessa coordinata x per tutti i punti
            y=[[plane["y"] - half_size, plane["y"] + half_size], [plane["y"] - half_size, plane["y"] + half_size]],
            z=[[-plane["height"]/2, -plane["height"]/2], [plane["height"]/2, plane["height"]/2]],
            colorscale=[[0, plane["color"]], [1, plane["color"]]],
            showscale=False,
            name=plane["name"],
            hovertext=f"<b>{plane['name']}</b><br>Position: ({plane['x']}, {plane['y']}, 0)<br>Size: {plane['size']}x{plane['height']} m",
            hoverinfo="text"
        ))
        # Aggiungi annotazione per il nome della stazione
        fig.add_trace(go.Scatter3d(
            x=[plane["x"]],
            y=[plane["y"] + half_size + 0.1],  # Posiziona il testo leggermente sopra il piano
            z=[plane["height"] / 2],
            mode="text",
            text=[plane["name"]],
            textfont=dict(size=14, color="black"),
            showlegend=False
        ))
    
        # Definisci un segmento 3D che va dal basso verso l'alto (lungo z)
    segment = {
        "name": "Upward Segment",
        "x": [-0.5, 0.5],  # Attraversa le stazioni 1, 2, e 3
        "y": [0.1, 0.1],   # Linea retta lungo y
        "z": [-0.4, 0.3],      # Dal basso (-1) verso l'alto (1)
        "color": "red",
        "width": 4
    }

    # Aggiungi il segmento 3D
    fig.add_trace(go.Scatter3d(
        x=segment["x"],
        y=segment["y"],
        z=segment["z"],
        mode="lines",
        line=dict(color=segment["color"], width=segment["width"]),
        name=segment["name"]
    ))

# Calcola i punti di intersezione tra il segmento e i piani
    crossing_points = []
    for plane in planes:
        # Parametri del segmento
        x_start, x_end = segment["x"]
        y_start, y_end = segment["y"]
        z_start, z_end = segment["z"]

        # Parametri del piano (x è costante per ogni piano)
        x_plane = plane["x"]

        # Calcola il parametro t per l'intersezione in x
        if x_start != x_end:
            t = (x_plane - x_start) / (x_end - x_start)
            if 0 <= t <= 1:  # L'intersezione è all'interno del segmento
                y_cross = y_start + t * (y_end - y_start)
                z_cross = z_start + t * (z_end - z_start)
                crossing_points.append((x_plane, y_cross, z_cross))

    # Aggiungi i punti di intersezione come markers
    for i, point in enumerate(crossing_points):
        fig.add_trace(go.Scatter3d(
            x=[point[0]],
            y=[point[1]],
            z=[point[2]],
            mode="markers",
            marker=dict(size=8, color="gold"),
            name=f"Crossing Point {i+1}"
        ))

    # Aggiungi un punto all'origine
    fig.add_trace(go.Scatter3d(
        x=[0], y=[0], z=[0],
        mode="markers",
        marker=dict(size=5, color="black"),
        name="Origin"
    ))

    # Imposta il layout per la visualizzazione 3D
    fig.update_layout(
        title="3D Interactive Detector Layout (Transparent Planes)",
        scene=dict(
            xaxis_title="x (m)",
            yaxis_title="y (m)",
            zaxis_title="z (m)",
            aspectmode="data"
        ),
        margin=dict(l=0, r=0, b=0, t=30)
    )

    fig.show()

plot_3d_transparent_planes()
```
