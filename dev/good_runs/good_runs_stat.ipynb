```python
%reload_ext autoreload
%autoreload 2

from glob import glob
from muraves_lib import run_manager
import pandas as pd


parsed_files_path = '/data/PARSED'
hodoscope = 'NERO'
version = 'v0'

path = f'{parsed_files_path}/{hodoscope}/{version}/'
file_list = glob(f"{path}/*.json")
print(f"Found {len(file_list)} files in {path}")


```

```python
import pandas as pd

result_df = pd.DataFrame(columns=["run", "parsing_failed", "has_less_events", "missing_bit_positivity", "missing_bit_notrecoverable"])

for file in file_list:
    run = file.split('/')[-1].split('.')[0].split('run')[-1]
    result = run_manager.read_summary_file(file)

    # Append a new row for each file
    result_df.loc[len(result_df)] = {
        "run": run,
        "parsing_failed": result.is_parsing_failed,
        "has_less_events": result.has_less_events,
        "missing_bit_positivity": result.missing_bit_positivity,
        "missing_bit_notrecoverable": result.missing_bit_notrecoverable
    }

print(result_df)

#adding a colims to summurise the runs that passed the missing bit test: True if  missing_bit_positivity ==False or (missing_bit_positivity == True and missing_bit_notrecoverable == False)

result_df["missing_bit_passed"] = (result_df["missing_bit_positivity"] == False) | ( (result_df["missing_bit_positivity"] == True) & (result_df["missing_bit_notrecoverable"] == False) )
```

```python
total_runs = len(result_df)

parsing_failed_count = len(result_df.query("parsing_failed == True and missing_bit_test_passed == True"))
print(f"Files that failed to parse: {parsing_failed_count} out of {total_runs}")
parse_ok_total = total_runs - parsing_failed_count
print(f"New total -> files that parsed successfully: {parse_ok_total} out of {total_runs}") 
has_less_events_count = len(result_df.query('has_less_events == True'))
relative_total_has_less_events = has_less_events_count / parse_ok_total * 100
print(f"Files that have less events: {has_less_events_count} out of {parse_ok_total} ({relative_total_has_less_events:.2f}%)")
missing_bit_positivity_count = len(result_df.query('missing_bit_positivity == True'))
relative_total_missing_bit_positivity = missing_bit_positivity_count / parse_ok_total * 100
print(f"Files that have missing bit positivity: {missing_bit_positivity_count} out of {parse_ok_total} ({relative_total_missing_bit_positivity:.2f}%)")
missing_bit_notrecoverable_count = len(result_df.query('missing_bit_test_passed == False'))
relative_total_missing_bit_notrecoverable = missing_bit_notrecoverable_count / parse_ok_total * 100
print(f"Files that have missing bit not recoverable: {missing_bit_notrecoverable_count} out of {parse_ok_total} ({relative_total_missing_bit_notrecoverable:.2f}%)")

#List of runs that pass all checks
runs_basic_checks = result_df.query('parsing_failed == False and has_less_events == False')
runs_missing_bit_check = runs_basic_checks.query('(missing_bit_test_passed == True)')

print(f"Runs that pass all checks: {len(runs_missing_bit_check)} out of {parse_ok_total} ({len(runs_missing_bit_check)/parse_ok_total*100:.2f}%)")
#print(runs_missing_bit_check['run'].tolist())
#sorting the runs that pass all checks by run number
runs_missing_bit_check_sorted = runs_missing_bit_check.sort_values(by='run')
print(runs_missing_bit_check_sorted['run'].tolist())

```
